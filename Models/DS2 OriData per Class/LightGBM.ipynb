```python
import glob as g
import numpy as np
import random as r
import cv5_Ori as c5
from lightgbm import LGBMClassifier
```

```python
file_index=[i for i in range(0, 17, 2)]
file_list=g.glob('C:\\Users\\SEO\\Desktop\\Makenaide\\class data csv\\*')

for idx in file_index :
    
    class_NG_data=np.loadtxt(file_list[idx], delimiter=',', dtype=np.float32)
    class_OK_data=np.loadtxt(file_list[idx+1], delimiter=',', dtype=np.float32)
    NG_size=len(class_NG_data) ; OK_size=len(class_OK_data)
    
    print(file_list[idx], end='\n\n')
    print('# of NG :',NG_size) ; print('# of OK :',OK_size) ; print('# of iters :',OK_size//NG_size)
    print()
    
    for i in range(0, OK_size//NG_size):
        
        data=np.vstack([class_NG_data, class_OK_data[i*NG_size:(i+1)*NG_size]])
        
        obj=c5.fold_5_cv(data, LGBMClassifier(objective='binary'), 'LighGBM')
        obj.train_model()
        
        del obj
        
    print('\n')
    
    del class_NG_data, class_OK_data
```
