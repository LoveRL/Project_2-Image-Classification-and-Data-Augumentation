```python
import numpy as np
import cv5 as c5
import glob as g
from lightgbm import LGBMClassifier

data=np.loadtxt('C:\\Users\\SEO\\Desktop\\Makenaide\\data\\구조4) Scikit auged 34552\\34552_sci_OK_data(without noise).csv', delimiter=',', dtype=np.float32)

model=LGBMClassifier(boosting_type='dart', objective='binary', learning_rate=0.02, num_leaves=120, subsample_for_bin=10000, class_weight={0:0.65, 1:0.35})

obj=c5.fold_5_cv(data, model, 'LightGBM')
obj.train_model()
```
